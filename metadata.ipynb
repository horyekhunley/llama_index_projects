```python
from llama_index.llms.ollama import Ollama

llm = Ollama(model="mistral-nemo:latest", request_timeout=60.0, temperature=0)

llm
```

```python
llm
```

```python
from llama_index.core.schema import MetadataMode
from llama_index.core.extractors import (SummaryExtractor, QuestionsAnsweredExtractor, TitleExtractor, KeywordExtractor, BaseExtractor)
from llama_index.extractors.entity import EntityExtractor
from llama_index.core.node_parser import TokenTextSplitter
```

```python
text_splitter = TokenTextSplitter(separator=" ", chunk_size=1000, chunk_overlap=256)
```

```python
extractor = [TitleExtractor(nodes=5, llm=llm), QuestionsAnsweredExtractor(questions=3, llm=llm)]
```

```python
transformations = [text_splitter] + extractor
```

```python
from llama_index.core import SimpleDirectoryReader

docs = SimpleDirectoryReader(input_files=["./papers/language_understanding_paper.pdf"]).load_data()
```

```python
docs[11]
```

```python
doc_front_pages = docs[0:1]
doc_content_pages = docs[10:11]
```

```python
full_docs = doc_front_pages + doc_content_pages
```

```python
from llama_index.core.ingestion import IngestionPipeline

pipeline = IngestionPipeline(transformations = transformations)

doc_nodes = pipeline.run(documents=full_docs)
```
